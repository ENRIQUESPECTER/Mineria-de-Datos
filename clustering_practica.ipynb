```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data = {
    "Horas de estudio": [1,2,1.5,8,9,8.5,3,3.5,4],
    "Calificaciones": [2,3,2.5,9,10,9.5,5,5.5,6]
}
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['Horas de estudio'], df['Calificaciones'])
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaciones")
plt.title("Datos de los estudiantes")
plt
```

```python
# Asumiendo que 'X' son tus datos preprocesados
wcss = []
for i in range(1, 9):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

# Graficar el método del codo
plt.plot(range(1, 9), wcss, c='gold')
plt.title('Método del Codo')
plt.xlabel('Número de clusters (k)')
plt.ylabel('WCSS (Inercia)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df)
```

```python
 #Mostrar los centroides
centroides = kmeans.cluster_centers_
centroides
```

```python
df['Cluster'] = kmeans.labels_
df
```

```python
#Visualizamos los centroides
plt.scatter(df['Horas de estudio'], df['Calificaciones'], c=df['Cluster'])
plt
```

```python
plt.scatter(centroides[:,0], centroides[:,1], c='black', s=200, alpha=0.5)
```
