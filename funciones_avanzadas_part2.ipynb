**EJERCICIOS**

**Une orders_a y orders_b en un solo DataFrame orders con índice reiniciado.**

**Ordena orders por order_date ascendente y, en caso de empate, por qty descendente.**

**Establece order_date como índice en orders y ordena por índice (cronológico).**

```python
orders = pd.concat([df,df2],ignore_index=True)
orders
```

```python
orders.sort_values(['order_date','qty'], ascending=[True,False])
```

```python
orders = orders.set_index('order_date')
orders.sort_index()
```

**Merge (enriquecer órdenes con clientes):**

**Haz un merge para añadir a orders las columnas del cliente**
**(customer_name, city, segment) usando customer_id (left join). Valida que sea many_to_one.**

```python
customers = (
    df3[['customer_id','customer_name','city','segment']]
)
customers
```

```python
fusion = orders.merge(customers, on="customer_id", how="left", validate="many_to_one")
fusion
```

**Merge (añadir producto y calcular total):**

**Haz un merge adicional con products para añadir product, category, unit_price y crea una columna line_total = qty * unit_price. Luego ordena por line_total descendente y muestra top 5.**

```python
products = df4
```

```python
fusion = orders.merge(products, on='product_id', how='left', validate='many_to_one')
fusion['line_total'] = fusion['qty'] * fusion['unit_price']

fusion.sort_values("line_total", ascending=False)
fusion
```
